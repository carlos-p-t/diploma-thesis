This notebook contains the visualization examples of how the best models perform their task (Billboard Recognition) in testing images.

```python
from ultralytics import YOLO
import os
import random
import cv2
from matplotlib import pyplot as plt
import requests
from PIL import Image
import numpy as np
```

```python
def view_results(results):
    result = results[0]  # Getting the first result
    image_with_boxes = result.plot()  # Plotting the image with bounding boxes
    # Display the image
    plt.imshow(cv2.cvtColor(image_with_boxes, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.show()
```

```python
images_path = '/home/t/troncoso1/Diploma/Testing_Images/'
images = os.listdir(images_path)
```

## Test Any Image

This section aims to test any given image in order to detect billboards using our best 2 models

```python
# Getting the image from the web

url = 'https://i.ytimg.com/vi/JqcZSltLMcE/maxresdefault.jpg'
response = requests.get(url, stream=True)
if response.status_code == 200:
    with open("downloaded_image.jpg", "wb") as file:
        file.write(response.content)
    print("Image downloaded successfully.")
else:
    print("Failed to download image.")
```

```python
# Loading the image

img = Image.open("downloaded_image.jpg")
plt.imshow(img)
plt.axis('off')
plt.show()
```

```python
# Fine Tuning YOLOv8-based Model (7th Fine Tuning Training)

fine_tuned_model = YOLO('/home/t/troncoso1/Diploma/fine_tuning_runs/detect/train7/weights/best.pt')
results3 = fine_tuned_model.predict(source='downloaded_image.jpg', imgsz=1280, conf=0.5)
view_results(results3)
```

```python
# Fine Tuning YOLO11-based Model (20th Fine Tuning Training)

fine_tuned_model2 = YOLO('/home/t/troncoso1/Diploma/fine_tuning_runs/detect/train20/weights/best.pt')
results4 = fine_tuned_model2.predict(source='downloaded_image.jpg', imgsz=1280, conf=0.5)
view_results(results4)
```
